##### FEATURE SELECTION

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
x = concat2.drop(['company'],axis=1)
y = concat2.company
```

METHOD 1 : RFE

```python
rfe_selector = RFE(estimator=RandomForestClassifier(),n_features_to_select = 2, step = 1)
m=rfe_selector.fit(x, y)
x.columns[m.get_support()]
print("Num Features: %s" % (m.n_features_))
print("Selected Features: %s" % (m.support_))
print("Feature Ranking: %s" % (m.ranking_))
```

```python
concat2
```

METHOD 2 : SFM

```python
SFM = SelectFromModel(estimator=RandomForestClassifier())
s=SFM.fit(x, y)
feature_idx = s.get_support()
feature_name = x.columns[feature_idx]
feature_name 
```

METHOD 3 : RFECV

```python
from sklearn.feature_selection import RFECV

rfecv = RFECV(
    estimator=RandomForestClassifier(),
    min_features_to_select=2,
    step=5,
    n_jobs=-1, 
    scoring="r2",
    cv=5,)
m= rfecv.fit(x, y)

x.columns[rfecv.support_]
```

Method 1 : input requests 2 features so this is what is given : company and job are the selectef features  
Method 2 : SFM selects 5 features : hierarchy_level1 and 2 (current and former); years_at_job; company2 (last company) and approx_work_xp (or approx age)  
Method 3 : Selects too many columns 

Keeping output from model 2 which seems the most reasonable

##### TEST TRAIN SPLIT

```python
from sklearn.model_selection import train_test_split
y = concat2.company
x = concat2[feature_name ] # features found thanks to SFM

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=5)
```

##### HYPERPARAMETER TUNING

METHOD 1 : GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
grid_search = GridSearchCV(RandomForestClassifier(random_state=0),
                           {
                              'n_estimators':np.arange(5,400,5)
                            
                            },cv=5, scoring="accuracy",verbose=1,n_jobs=-1 
                           ) 
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

METHOD 2 : RANDOM SEARCH

```python
rand_search = RandomizedSearchCV(RandomForestClassifier(random_state=0),
                           {
                              'n_estimators':np.arange(5,400,5)
                            
                            },cv=5, scoring="accuracy",verbose=1,n_jobs=-1
                           )
rand_search.fit(X_train,y_train)
```

```python
print(rand_search.best_params_)
print(rand_search.best_score_)
```

Method 1 : has a slightly better score with 81,8% and 120 n_estimators  
Method 2 : score of 81,25% with 325 n_estimators  

Using Method 1 for next steps

#### MODELS

RANDOM FOREST CLASSIFIER

```python
n_estimators = 120
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
def generate_results(model, predictions, name):
    cl_rep = classification_report(y_test, predictions)
    print("\nThe classification report for " + name + " is:", cl_rep, sep = "\n")
    cm_model = confusion_matrix(y_test, predictions)
    plt.figure(figsize = (8, 6))
    sns.heatmap(cm_model, annot = True, cmap = 'Blues', annot_kws = {'size': 15}, square = True)
    plt.title('Confusion Matrix for ' + name, size = 15)
    plt.xticks(size = 15)
    plt.yticks(size = 15)
    plt.show() 
```

```python
rfc = RandomForestClassifier(n_estimators=n_estimators)
rfc.fit(X_train, y_train)
y_pred_rfc = rfc.predict(X_test)
acc_rfc = rfc.score(X_test, y_test)
print('The accuracy of the Random Forest Classifier is:', acc_rfc * 100, '%')
```

```python
print('The accuracy of the Random Forest Classifier is:', acc_rfc * 100, '%')
generate_results(rfc, y_pred_rfc, 'Random Forest Classifier')
```

K NEIGHBOURS CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    print ('-------------', k, '-------------------')
    print(metrics.confusion_matrix(y_test, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.show()
```

Keeping value k= 4 for KNN

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print('The accuracy of the KNeighbors Classifier:', knn.score(X_test, y_test) * 100, '%')
generate_results(lr, y_pred_lr, 'KNeighbors Classifier')
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

score=[]
for max_depth in range(2,20):
    tree = DecisionTreeClassifier(max_depth = max_depth, random_state = 1)
    tree.fit(X_train,y_train)
    prediction=tree.predict(X_test)
    score.append(metrics.accuracy_score(prediction,y_test))
    print('with max_depth :', max_depth)
    print("The accuracy of the Decision Tree is","{:.3f}".format(metrics.accuracy_score(prediction,y_test)))
```

```python
max_depth_range = range(2,20)
plt.plot(max_depth_range, score)
plt.xlabel('Value of max_depth for DecisionTree')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of max_depth of DecisionTree')
plt.show()
```

```python
score=[]
for min_sample in range(2,20):
    tree = DecisionTreeClassifier(max_depth = 2, random_state = 1, min_samples_split = min_sample)
    tree.fit(X_train,y_train)
    prediction=tree.predict(X_test)
    score.append(metrics.accuracy_score(prediction,y_test))
    print('with min_samples_split :',min_sample )
    print("The accuracy of the Decision Tree is","{:.3f}".format(metrics.accuracy_score(prediction,y_test)))
```
